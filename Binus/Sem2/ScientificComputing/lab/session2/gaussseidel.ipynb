```python
import numpy as np
```

```python
def gauss_seidel(A, b, x0=None, tol=1e-6, max_iter=100, verbose = True):
    n = len(b)

    #initial guess (all zero)
    if x0=None:
        x = np.zeros(n, dtype=float)
    else:
        x = np.array(n, dtype=float)

    #check diagonally dominant
    for i in range(n):
        diag = abs(A[i,i])
        off_diag=sum(abs(A[i,j]) for j in range(n) if j!=i)
        if diag<=off_diag:
            print("not diagonally dominant")
            return
        
    if verbose:
        print(f"{'Iter':>4} | {'x':}")
```

```julia
import numpy as np

def gauss_seidel(A, b, x0=None, tol=1e-6, max_iter=100, verbose=True):
    """
    Solves Ax = b using the Gauss-Seidel iterative method.
    
    Parameters:
    -----------
    A : 2D array (n x n) - Coefficient matrix
    b : 1D array (n,)    - Right-hand side vector
    x0 : 1D array (n,)   - Initial guess (default: zeros)
    tol : float           - Stopping tolerance
    max_iter : int        - Maximum iterations
    verbose : bool        - Print iteration history
    
    Returns:
    --------
    x : 1D array          - Solution vector
    iters : int           - Number of iterations performed
    """
    n = len(b)
    
    # Default initial guess: all zeros
    if x0 is None:
        x = np.zeros(n, dtype=float)
    else:
        x = np.array(x0, dtype=float)
    
    # Check diagonal dominance (warning only)
    for i in range(n):
        diag = abs(A[i, i])
        off_diag_sum = sum(abs(A[i, j]) for j in range(n) if j != i)
        if diag <= off_diag_sum:
            print(f"⚠️  WARNING: Row {i} is NOT strictly diagonally dominant.")
            print(f"   |a_ii| = {diag}, sum of others = {off_diag_sum}")
            print("   Convergence is NOT guaranteed!\n")
    
    if verbose:
        print(f"{'Iter':>4} | {'x':>40} | {'Max Change':>12}")
        print("-" * 65)
        print(f"{0:>4} | {str(np.round(x, 6)):>40} | {'-':>12}")
    
    for k in range(1, max_iter + 1):
        x_old = x.copy()
        
        # Update each variable using the MOST RECENT values
        for i in range(n):
            # Sum of terms with j < i (use NEW values)
            sum_new = sum(A[i, j] * x[j] for j in range(i))
            
            # Sum of terms with j > i (use OLD values)
            sum_old = sum(A[i, j] * x_old[j] for j in range(i + 1, n))
            
            x[i] = (b[i] - sum_new - sum_old) / A[i, i]
        
        # Check convergence: maximum absolute change
        max_change = np.max(np.abs(x - x_old))
        
        if verbose:
            print(f"{k:>4} | {str(np.round(x, 6)):>40} | {max_change:>12.2e}")
        
        if max_change < tol:
            if verbose:
                print(f"\n✅ Converged in {k} iterations.")
            return x, k
    
    print(f"\n❌ Did not converge within {max_iter} iterations.")
    return x, max_iter


# ==================== EXAMPLE ====================
if __name__ == "__main__":
    # Coefficient matrix
    A = np.array([
        [4, 1, 1],
        [1, 5, 1],
        [1, 1, 6]
    ], dtype=float)
    
    # Right-hand side
    b = np.array([6, 7, 8], dtype=float)
    
    # Solve
    solution, iterations = gauss_seidel(A, b, tol=1e-6, max_iter=20)
    
    print(f"\n{'='*50}")
    print(f"Solution:     {np.round(solution, 6)}")
    print(f"True solution: [1. 1. 1.]")
    print(f"Iterations:   {iterations}")
    
    # Verify: compute Ax and compare with b
    residual = A @ solution - b
    print(f"Residual (Ax-b): {np.round(residual, 10)}")
```

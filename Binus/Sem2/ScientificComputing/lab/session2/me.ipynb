```python
import numpy as np

```

```python
def gaussSeidel(A, b, x0, max_iter, tol=0.000001):
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    n = len(b)

    #diagonal dominance check
    diag = np.abs(np.diag(A))
    off_diag = np.sum(np.abs(A), axis=1)-diag
    if not np.all(diag>=off_diag):
        print("not diagonally dominant, convergence is not guaranteed\n")

    if x0 is None:
        x=np.zeros_like(b,dtype=float)
    else:
        x=np.array(x,dtype=float)

    for i in range(max_iter):
        x_old=x.copy()
        for j in range(n):
            s1=np.dot(A[j,:j], x[:j])
            s2=np.dot(A[j, j+1:], x_old[j+1:])

            x[j]=(b[j]-s1-s2) / A[j,j]
        error = np.linalg.norm(x - x_old, ord=np.inf)
        if error < tol:
            print(f"Success: Converged to tolerance {tol} after {i+1} iterations.")
            return x
        
    print("Failed to converge within maximum iterations.")
    return x

```

```python
# Define the coefficient matrix A
A = [
    [4,  1, -1],
    [2,  5,  2],
    [1,  2,  4]
]

# Define the constants vector b
b = [8, 3, 11]

# Run the solver
solution = gaussSeidel(A, b, x0=None, max_iter=1000)

print("\nSolution Vector (x):", solution)
```

```python
import numpy as np

# Data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 6])

def least_squares(x, y):
    n = len(x)
    
    # Calculate all the sums
    sum_x  = np.sum(x)
    sum_y  = np.sum(y)
    sum_xy = np.sum(x * y)
    sum_x2 = np.sum(x**2)
    
    # Calculate a1 and a0
    a1 = (n*sum_xy - sum_x*sum_y) / (n*sum_x2 - sum_x**2)
    a0 = np.mean(y) - a1 * np.mean(x)
    
    return a0, a1

a0, a1 = least_squares(x, y)
print(f"Slope (a1):     {a1:.4f}")
print(f"Intercept (a0): {a0:.4f}")
print(f"Line equation:  y = {a0:.4f} + {a1:.4f}x")

# Predict values
print("\nPredicted vs Actual:")
for i in range(len(x)):
    predicted = a0 + a1*x[i]
    print(f"x={x[i]}: predicted={predicted:.2f}, actual={y[i]}, residual={y[i]-predicted:.2f}")
```

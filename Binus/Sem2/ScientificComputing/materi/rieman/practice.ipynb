```python
import numpy as np

def f(x):
    return x**2

def riemann(f,a,b,n,method='midpoint'):
    h=(b-a)/n
    total = 0

    for i in range(n):
        x_left = a + i * h

        if method == 'left':
            height=f(x_left)
        elif method == 'right':
            height=f(x_left+h)
        elif method == 'midpoint':
            height=f(x_left+h/2)
        total += height*h
    return total

```

```python
import numpy as np

def f(x):
    return x**2   # Function to integrate

def riemann(f, a, b, n, method='midpoint'):
    h = (b - a) / n                    # Width of each rectangle
    total = 0

    for i in range(n):
        x_left = a + i * h             # Left edge of current slice

        if method == 'left':
            height = f(x_left)
        elif method == 'right':
            height = f(x_left + h)
        elif method == 'midpoint':
            height = f(x_left + h/2)
        
        total += height * h            # Area = height × width

    return total

# Test all three methods
a, b, n = 0, 2, 4
exact = 8/3  # Real answer

print(f"Exact answer:    {exact:.6f}")
print(f"Left Riemann:    {riemann(f, a, b, n, 'left'):.6f}")
print(f"Right Riemann:   {riemann(f, a, b, n, 'right'):.6f}")
print(f"Midpoint Riemann:{riemann(f, a, b, n, 'midpoint'):.6f}")
```

```python
import numpy as np
```

```python
x = [
    [5,-2,2],
    [4,6,-1],
    [-3,1,5],
]

y = [7,8,4]
```

```python
def check_dominant(x):
    x = np.array(x)

    diag = np.diag(np.abs(x))

    not_diag = np.sum(np.abs(x), axis=1) - diag

    if np.all(diag > not_diag):
        return True
    else:
        return False
```

```python
def gauss_seidel(x,y,e=0.01,n=50):
    if not check_dominant(x):
        print("Not dominant")
        return
    
    x = np.array(x)
    y = np.array(y)

    diag = np.diag(x)

    x = -x

    np.fill_diagonal(x,0)
    mtx_old = np.zeros(x[1].shape)

    for i in range(n):
        mtx_new = np.array(mtx_old)
        
        for j, row in enumerate(x):
            mtx_new[j] = (y[j] + np.dot(mtx_new,row)) / diag[j]

        print(f"iterasi ke {i+1} adalah: ")
        print(mtx_new)

        euc = np.sqrt(np.dot(mtx_new - mtx_old, mtx_new - mtx_old))

        if euc < e:
            print ("iterasi selesai")
            return
        
        mtx_old = mtx_new

gauss_seidel(x,y)
```

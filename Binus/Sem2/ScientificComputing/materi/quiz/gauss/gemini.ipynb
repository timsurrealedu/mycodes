```python
import numpy as np

def gauss_seidel(A, b, x0=None, tol=1e-8, max_iter=1000):
    """
    Solves the system of linear equations Ax = b using the Gauss-Seidel iterative method.
    
    Parameters:
        A (list or np.ndarray): The coefficient matrix (n x n).
        b (list or np.ndarray): The constant vector (n).
        x0 (list or np.ndarray, optional): Initial guess vector. Defaults to zeros.
        tol (float): Tolerance for the stopping criterion.
        max_iter (int): Maximum number of iterations allowed.
        
    Returns:
        np.ndarray: The solution vector x.
    """
    # Convert inputs to numpy arrays for reliable math operations
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    n = len(b)
    
    # 1. FEATURE: Diagonal Dominance Check
    # Gauss-Seidel is only guaranteed to converge if the matrix is strictly diagonally dominant 
    # (or symmetric positive definite). We provide a helpful warning if it isn't.
    diag = np.abs(np.diag(A))
    off_diag = np.sum(np.abs(A), axis=1) - diag
    if not np.all(diag > off_diag):
        print("Warning: Matrix A is not strictly diagonally dominant. Convergence is not guaranteed.")
        
    # 2. FEATURE: Initial Guess Handling
    if x0 is None:
        x = np.zeros_like(b, dtype=float)
    else:
        x = np.array(x0, dtype=float)
        
    # 3. FEATURE: Iteration Limit (Prevents infinite loops)
    for k in range(max_iter):
        x_old = x.copy()
        
        # Core Gauss-Seidel Loop
        for i in range(n):
            # We use x[:i] (the NEW updated values) and x_old[i+1:] (the OLD values)
            # This immediate use of new values is what separates Gauss-Seidel from Jacobi.
            s1 = np.dot(A[i, :i], x[:i])
            s2 = np.dot(A[i, i+1:], x_old[i+1:])
            
            # The core mathematical update formula
            x[i] = (b[i] - s1 - s2) / A[i, i]
            
        # 4. FEATURE: Convergence/Tolerance Check
        # Calculate the maximum absolute difference between the old and new x values
        error = np.linalg.norm(x - x_old, ord=np.inf)
        if error < tol:
            print(f"Success: Converged to tolerance {tol} after {k+1} iterations.")
            return x
            
    print(f"Warning: Stopped after reaching maximum iterations ({max_iter}) without hitting tolerance.")
    return x
```

```python
# Define the coefficient matrix A
A = [
    [4,  1, -1],
    [2,  5,  2],
    [1,  2,  4]
]

# Define the constants vector b
b = [8, 3, 11]

# Run the solver
solution = gauss_seidel(A, b)

print("\nSolution Vector (x):", solution)
```

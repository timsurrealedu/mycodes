```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.15, 0.32, 0.48, 0.71, 0.89, 1.03, 1.25, 1.41, 1.58, 1.76,
              1.92, 2.11, 2.35, 2.48, 2.67, 2.81, 3.04, 3.19, 3.37, 3.55])

y = np.array([1.12, 1.58, 2.01, 2.76, 3.10, 3.54, 4.02, 4.45, 4.91, 5.36,
              5.82, 6.31, 6.91, 7.12, 7.65, 8.01, 8.56, 8.98, 9.43, 9.92])

# Step 2: hitung semua komponen
n      = len(x)          # 20
sum_x  = np.sum(x)       # Σx
sum_y  = np.sum(y)       # Σy
sum_xy = np.sum(x * y)   # Σxy
sum_x2 = np.sum(x**2)    # Σx²

print(f"n      = {n}")
print(f"Σx     = {sum_x:.4f}")
print(f"Σy     = {sum_y:.4f}")
print(f"Σxy    = {sum_xy:.4f}")
print(f"Σx²    = {sum_x2:.4f}")

# Step 3: hitung m (slope)
m = (n*sum_xy - sum_x*sum_y) / (n*sum_x2 - sum_x**2)

# Step 4: hitung c (intercept)
c = np.mean(y) - m * np.mean(x)

print(f"\nm (slope)     = {m:.4f}")
print(f"c (intercept) = {c:.4f}")
print(f"\nPersamaan garis: y = {m:.4f}x + {c:.4f}")

# Step 5: plot
x_line = np.linspace(min(x), max(x), 100)   # 100 titik untuk garis
y_line = m * x_line + c                      # hitung y untuk setiap titik

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Original Data', zorder=5)
plt.plot(x_line, y_line, color='red', label=f'y = {m:.4f}x + {c:.4f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Least Square Regression')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('least_squares.png')
plt.show()
```

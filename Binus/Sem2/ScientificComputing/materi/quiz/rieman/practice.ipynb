```python
import numpy as np
import matplotlib.pyplot as plt

# ── the function and interval ────────────────────────────────────────────────
f = lambda x: np.sin(x) + 1
a, b = 0, np.pi
n = 10
dx = (b - a) / n

# Create continuous x values to plot the smooth curve smoothly
x_curve = np.linspace(a, b, 200)
y_curve = f(x_curve)

# Define methods and plot configurations
methods = ["left", "right", "midpoint"]
colors = ["#ff7f0e", "#2ca02c", "#d62728"]

# Set up a 1x3 side-by-side subplot layout
fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

for ax, method, color in zip(axs, methods, colors):
    # 1. Plot the smooth continuous function line
    ax.plot(x_curve, y_curve, color="blue", linewidth=2, label="$f(x) = \sin(x) + 1$")
    
    # 2. Compute indices and evaluate x positions for rectangles
    i = np.arange(n)
    if method == "left":
        xs = a + i * dx
        align = "edge"  # Left edge of bar maps to xs
    elif method == "right":
        xs = a + (i + 1) * dx
        align = "edge"  # Shifting manually below via width parameter
    elif method == "midpoint":
        xs = a + (i + 0.5) * dx
        align = "center" # Center of bar maps to xs

    ys = f(xs)
    
    # 3. Handle specific bar positioning alignments 
    bar_xs = xs if method != "right" else xs - dx
    
    # 4. Render the Riemann approximation rectangles
    ax.bar(bar_xs, ys, width=dx, align=align, alpha=0.3, 
           color=color, edgecolor=color, linewidth=1.5, label=f"{method.capitalize()} Sum")
    
    # 5. Plot the contact reference dots on the curve
    ax.scatter(xs, ys, color=color, zorder=3, s=30)
    
    # Formatting details
    ax.set_title(f"{method.capitalize()} Method (n={n})")
    ax.set_xlabel("x")
    ax.set_xlim(a - 0.2, b + 0.2)
    ax.grid(True, linestyle="--", alpha=0.5)

axs[0].set_ylabel("y")
plt.tight_layout()
plt.show()
```

```python
import numpy as np

def f(x):
    return -3*x**4 - 7 * np.sin(x)**3 - 11 * x**2 + 6

def riemann(f,a,b,n,method):
    h=(b-a)/n
    total = 0

    for i in range(n):
        x_left = a + i * h
        if method == 'left':
            height = f(x_left)
        elif method == 'right':
            height = f(x_left+h)
        elif method == 'midpoint':
            height = f(x_left+h/2)
        elif method == 'trapezoidal':
            height = (f(x_left)+f(x_left+h))/2
        else:
            print("not a valid method")
            return None
        
        total += height * h
    return total

print(f"left: {riemann(f,-3,3,3000,'left')}")
print(f"right: {riemann(f,-3,3,3000,'right')}")
print(f"midpoint: {riemann(f,-3,3,3000,'midpoint')}")
print(f"trapezoidal: {riemann(f,-3,3,3000,'trapezoidal')}")
```

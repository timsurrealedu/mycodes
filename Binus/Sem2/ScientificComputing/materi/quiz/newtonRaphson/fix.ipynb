```python
def f(x):
    return x**2 - 2  # Only define f, no f' needed!

def numerical_derivative(f, x, h=1e-7):
    return (f(x + h) - f(x)) / h  # Approximation of f'(x)

def newton_raphson_auto(f, x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        fpx = numerical_derivative(f, x)  # Auto derivative!
        
        if abs(fpx) < 1e-12:
            print("Derivative is zero. Method fails.")
            return None
        
        x_new = x - fx / fpx
        error = abs(x_new - x)
        print(f"Iter {i}: x = {x_new:.10f}, error = {error:.2e}")
        
        if error < tol:
            print(f"\n✅ Root: {x_new:.10f}")
            return x_new
        x = x_new

newton_raphson_auto(f, x0=1.0)
```

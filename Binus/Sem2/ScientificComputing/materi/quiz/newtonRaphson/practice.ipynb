```python
def f(x):
    return x**2 - 2

def deriv(f,x,h=1e-7):
    return (f(x+h)-f(x))/h

def newtonRaphson(f,x,tol=1e-6,max_iter=1000):
    for i in range(max_iter):
        fx=f(x)
        fpx=deriv(f,x)
        if abs(fpx)<1e-12:
            return None
        x_new = x - fx/fpx
        error = abs(x_new-x)
        print(f"iterations {i}, x = {x_new}")
        if error < tol:
            print(f"final is {x_new}")
            return None
        x=x_new

newtonRaphson(f,1.0)
```

```python
def f(x):
    return 2*x**3 - 5*x**2 + 4 * x - 1

def deriv(f,x,h=1e-6):
    return (f(x+h)-f(x))/h

def newtonRaphson(f,x,tol,max_iter):
    for i in range(max_iter):
        fx=f(x)
        fpx=deriv(f,x)
        if abs(fpx)<1e-12:
            print("derivative is zero, cannot proceed")
            return None
        x_new = x - fx/fpx
        error = abs(x_new-x)
        print(f"iteration {i}: x = {x_new}, error = {error}")
        if error<tol:
            print(f"final result reached after {i} iterations, x = {x_new}")
            return x_new
        x=x_new

    print("reached maximum iterations without converging")
    return None

newtonRaphson(f,5,0.05,10)
```

```python
def f(x):
    return x**2 - 2

def deriv(f,x,h=1e-7):
    return (f(x+h)-f(x))/h

def newtonRaphson(f,x,tol=1e-6,max_iter=1000):
    for i in range(max_iter):
        fx=f(x)
        fpx=deriv(f,x)
        if abs(fpx)<1e-12:
            return None
        x_new = x - fx/fpx
        error = abs(x_new-x)
        print(f"iterations {i}, x = {x_new}")
        if error < tol:
            print(f"final is {x_new}")
            return None
        x=x_new

newtonRaphson(f,1.0)
```

```python
def f(x):
    return x**2 -2

def derivative(f, x, h=1e-7):
    return (f(x+h)-f(x))/h

def newtonRaphson(f,x0,tol=1e-6,max_iter=1000):
    x=x0
    for i in range(max_iter):
        fx=f(x)
        fpx=derivative(f,x)
        if abs(fpx)< 1e-12:
            print("derivative is zero so it doesnt work\n")
            return None
        
        x_new=x-fx/fpx
        error = abs(x_new-x)
        print(f"iter {i}: x = {x_new:.10f}, error = {error:.2e}")

        if error<tol:
            print(f"root: {x_new:.10f}")
            return x_new
        x=x_new

def f(x): 
    return x**3 + 3

newtonRaphson(f,1.0)
```

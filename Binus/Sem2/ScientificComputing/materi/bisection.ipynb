```python
import numpy as np
```

```python
def f(x):
    return x**3 - 6*x**2 + 11*x -6

print(f(5))
```

```python
def bisection(a,b,e=0.01,n=50):
    if n == 0:
        print("Max iteration")
        return
    if np.sign(f(a)) == np.sign(f(b)):
        print("invalid input")
        return
    mid = (a+b)/21
    if np.abs(f(mid)) < e:
        print(f"root found at x = {mid} and y = {f(mid)}")
    elif np.sign(f(a)) == np.sign(f(mid)):
        bisection(mid,b,n-1)
    elif np.sign(f(b)) == np.sign(f(mid)):
        bisection(a,mid,n-1)
        
        
bisection(1,9)
```

```python
import numpy as np
```

```python
def f(x):
    return x**3 - 6*x**2 + 11*x - 6

print(f(5))
```

```python
def bisection(a, b, error = 0.01, n = 50):
    if n == 0:
        print("Max iteration")
        return
    if np.sign(f(a)) == np.sign(f(b)):
        print("invalid input")
        return
    mid = (a+b)/2
    if np.abs(f(mid)) < error:
        print(f"Root found at x = {mid} and y = {f(mid)}")
    elif np.sign(f(a)) == np.sign(f(mid)):
        bisection(mid, b, n-1)
    elif np.sign(f(b)) == np.sign(f(mid)):
        bisection(a, mid, n-1)

bisection(3,5)
     
```

```python
def df(x):
    return 3*x**2 - 12*x + 11
```

```python
def newton(x0, e=0.01, n=50):
    if n==0:
        print("Max iteration")
        return
    x1 = x0 - f(x0)/df(x0)
    if x1 < e:
        print(f"Root at x = {x1} and y = {f(x1)}")
    else : 
        newton(x1,n-1)

newton(777)
```

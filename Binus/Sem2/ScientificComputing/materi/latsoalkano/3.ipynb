```python
import numpy as np
def f(x):
    return -3*x**4 - 7*np.sin(x)**3 - 11*x**2 + 6

def riemann(a,b,n,method):
    h=(b-a)/n
    total = 0
    for i in range(n):
        x_left = a + i * h
        if method == 'left':
            height = f(x_left)
        elif method == 'right':
            height = f(x_left+h)
        elif method == 'midpoint':
            height = f(x_left + h/2)
        elif method == 'trapezoidal':
            height = (f(x_left)+f(x_left+h))/2
        else:
            return
        total += height*h
    return total

print(f"left: {riemann(-3,3,3000,'left')}")
print(f"right: {riemann(-3,3,3000,'right')}")
print(f"midpoint: {riemann(-3,3,3000,'midpoint')}")
print(f"trapezoidal: {riemann(-3,3,3000,'trapezoidal')}")
```

```python
def f(x):
    return 2*x**3 - 5*x**2 + 4*x - 1

def deriv(f,x,h=1e-7):
    return (f(x+h)-f(x))/h

def newton(f,x,tol,max_iter):
    for i in range(max_iter):
        fx = f(x)
        fpx = deriv(f,x)
        if fpx<1e-12:
            return
        
        x_new = x - fx/fpx
        error = abs(x_new-x)
        print(f"iterations {i+1}: {x_new}")
        x=x_new
        if error<tol:
            print("done")
            return x
        
newton(f,5,0.05,10)
```

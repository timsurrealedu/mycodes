```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.15, 0.32, 0.48, 0.71, 0.89, 1.03, 1.25, 1.41, 1.58, 1.76, 1.92, 2.11, 2.35, 2.48, 2.67, 2.81, 3.04, 3.19, 3.37, 3.55])
y = np.array([1.12, 1.58, 2.01, 2.76, 3.10, 3.54, 4.02, 4.45, 4.91, 5.36, 5.82, 6.31, 6.91, 7.12, 7.65, 8.01, 8.56, 8.98, 9.43, 9.92])

def leastSquare(x,y):
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_xy = np.sum(x*y)
    sum_x2 = np.sum(x**2)

    n = len(x)

    m = (n*sum_xy - sum_x*sum_y)/(n*sum_x2 - sum_x**2)
    c = np.mean(y) - m*np.mean(x)

    print(f"m = {m}")
    print(f"c = {c}")
    x_line = np.linspace(min(x),max(x),1000)
    y_line = m*x_line + c


    plt.scatter(x,y,color='blue',label='original data')
    plt.plot(x_line,y_line,color='red',label='plotline')
    plt.grid()
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()


leastSquare(x,y)

```

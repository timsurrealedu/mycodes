```python
import numpy as np
```

```python
x = [
    [10,-2,1,1],
    [1,9,-3,2],
    [2,-1,11,-2],
    [-1,2,-3,12]
]

y = [17,-14,28,-33]
```

```python
def dominanceCheck(x):
    x = np.array(x)
    diag = np.diag(np.abs(x))
    not_diag = np.sum(np.abs(x),axis=1) - diag
    if np.all(diag>not_diag):
        return True
    return False

def gaussSeidel(x,y,e=0.022,max_iter=15):
    if not dominanceCheck(x):
        print("error")
        return None
    x = np.array(x,dtype=float)
    y = np.array(y,dtype=float)
    diag = np.diag(x)
    x = -x
    np.fill_diagonal(x,0)
    mtx_old = np.zeros(x[1].shape)
    for i in range(max_iter):
        mtx_new=np.array(mtx_old)
        for j, row in enumerate(x):
            mtx_new[j]=(y[j]+np.dot(mtx_new,row))/diag[j]
        print(f"iterations {i+1}: {mtx_new}")

        euc = np.sqrt(np.dot(mtx_new-mtx_old,mtx_new-mtx_old))
        if euc<e:
            print(f"done: {mtx_new}")
            return mtx_new
        
        mtx_old=mtx_new
    
gaussSeidel(x,y)
```
